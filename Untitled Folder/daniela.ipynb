```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
df=pd.read_csv('Datos_Corregidos_totales_2.5.csv')
columnsEstaciones=[' San Cristobal ',' Carvajal - Sevillana ', ' Guaymaral ', ' Puente Aranda ',
       ' Tunal    ', ' Fontibon ', ' Centro de Alto Rendimiento ',
       ' Usaquen ', ' Suba     ', ' Las Ferias ', ' Kennedy  ',
       ' MinAmbiente ']
```

```python
fecha=' Fecha y Hora              '
df[fecha]=pd.to_datetime(df[fecha], errors='coerce') #Convertir la fecha en  datatime
df[fecha].head() #Una muestra
```

```python
dtFrameYear=df.groupby(df[fecha].dt.year).mean() #Promedio por año de todas las estaciones
x=dtFrameYear.index
dtFrameYear
```

```python
for txt in columnsEstaciones:
    plt.plot(x,dtFrameYear[txt])
    plt.title(txt )
    plt.show()
```

```python
dtFrameHora=df.groupby(df[fecha].dt.hour).mean() #Promedio por hora
x=dtFrameHora.index
dtFrameHora
```

```python
for txt in columnsEstaciones:
    plt.plot(x,dtFrameHora[txt])
    plt.title(txt)
    plt.show()
```
