```python
#Importamos la libreria pandas
import pandas as pd
```

```python
#Lee el archivo movie.csv y lo convierte en un dataframe
movie=pd.read_csv('../Data/movie.csv')
```

```python
#Del data frame movie selecciona las columnas 'movie_title','title_year','content_rating','budget'
#Y guarda en nuevo dataframe en la variable movie4
movie4=movie[['movie_title','title_year','content_rating','budget']] 

#Muestra de 5 filas de la cabezera movie4
movie4.head(5)
```

```python
# Este codigo hace ordenamiento multiple
# Ordena primero la columna 'title_year' en forma descendiente, seguido de la columna
# 'content_rating' de forma descendiente y por Ãºltimo la columna 'budget' de forma ascendente
# Al dataframe ordenado lo asignamos a la variable movi4_sorted
movie4_sorted=movie4.sort_values(['title_year','content_rating','budget'], 
                                 ascending=[False,False,True])
#Resultado, muestra de 5
movie4_sorted.head(5)
```

```python
# Del data frame movi4_sorted eliminaremos los duplicados que se tenga en las columnas 
# 'title_year' y 'content_rating' de forma unida. Para aclarar: elimina datos duplicados
# luego de la concatenaciÃ³n de esas dos columnas. (No elimina datos duplicados de forma 
# independiente de columnas sino de la uniÃ³n de esas dos columnas)

# El nuevo dataframe sin los duplicados lo guardamos en la variable dat 
dat=movie4_sorted.drop_duplicates(subset=['title_year','content_rating'])

# Una muestra de 10
dat.head(10)
```

```python

# Creamos un dataframe y definimos las columnas iniciales de las columnas 
# del dataframe movie
new_df=pd.DataFrame(columns=movie.columns)


# Leemos el archivo movie.csv como un dataframe y con el atributo 'chunksize' dividimos el 
# dataframe en multiples dataframes con mÃ¡ximo de filas de valor 5
# AsÃ­ que este for recorrer los dataframes y por cada ciclo 
# guarda el dataframe en una variable llamada movie: Representa un dataframe de mÃ¡ximo 5 filas

for movie in pd.read_csv('../Data/movie.csv',chunksize=5):
    # Agrupamos las filas por la columna 'movie_title', ahora esta columna se vuelve 
    # la columna Indixe
    # Luego de agruparlos, sacamos 'count()' (conteo) de las columnas restantes,
    # BÃ¡sicamente cuenta cuantas veces se repite cada dato en cada fila y el valor final
    # se vuelve el dato en esa columna con indixe de la columna 'movie_title'
    result=movie.groupby(['movie_title']).count()
    
    #Se concatena el dataframe 'result' con el dataframe 'result' para cada iteracciÃ³n
    new_df=pd.concat([new_df,result])
```
