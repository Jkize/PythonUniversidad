```python
college = pd.read_csv('../Data/college.csv')
different_cols = ['RELAFFIL', 'SATMTMID', 'CURROPER', 'INSTNM', 'STABBR']
col2 = college.loc[:, different_cols]
col2.head()
```

```python
col2.dtypes
```

```python
''' Con el método memory_usage,
observamos el uso de la memoria de
cada columna. (Bytes Consumidos)'''
original_mem = col2.memory_usage(deep=True)
original_mem
```

```python
'''No es necesario utilizar 64 bits para la
columna RELAFILL ya que sólo
contiene 0/1. Convirtamos esta
columna en un entero de 8 bits (1 byte)
con el método "astype".'''
col2['RELAFFIL'] = col2['RELAFFIL'].astype(np.int8)
col2.dtypes
```

```python
'''Para ahorrar aún más memoria, se
considerara cambiar los tipos de datos
de los objetos a categóricos si tienen
una cardinalidad razonablemente baja
(valores numéricos únicos).
Comprobemos primero el número de
valores únicos para ambas columnas
de objetos.'''
col2.select_dtypes(include=['object']).nunique()
```

```python
'''La columna STABBR es una buena
candidata para convertir a Categórica
ya que menos del uno por ciento de
sus valores son únicos.'''
col2['STABBR'] = col2['STABBR'].astype('category')
col2.dtypes
```

```python
'''Calculemos el uso de la memoria de
nuevo'''
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
#Finalmente se compara la memoria usada con la actualizada.
new_mem / original_mem
```

```python
movie = pd.read_csv('../Data/movie.csv')
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.head() 
```

```python
'''1.- Usamos el método “nlargest” para seleccionar las mejores 100 peliculas por
imbd_score'''

movie2.nlargest(100, 'imdb_score').head()
```

```python
'''2.- Se encadena el método "nsmallest" para devolver las cinco películas de
menor presupuesto entre las 100 de mayor puntuación:'''
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

```python
movie = pd.read_csv('../Data/movie.csv')
movie2 = movie[['movie_title', 'title_year', 'imdb_score']]
```

```python
'''2.- Utiliza el método sort_values para ordenar el dataFrame por title_year. El
comportamiento por defecto ordena desde el más pequeño al más grande. Use
el parámetro “ascending” para invertir este comportamiento poniéndolo igual a
"True"'''
movie2.sort_values('title_year', ascending=False).head()
```

```python
'''3.- Fijate que solo se ordeno el año. Para ordenar multiples columnas al tiempo,
usamos una lista.'''
movie3 = movie2.sort_values(['title_year','imdb_score'], ascending=False)
movie3.head()
```

```python
'''4.- Ahora, usamos el método drop_duplicates para mantener sólo la primera
fila de cada año.'''
movie_top_year = movie3.drop_duplicates(subset='title_year') #Quito repetidos del año
movie_top_year.head()
```

```python
#Ordenar por titulo de mayor a menor, y luego el otro de menor a mayor
movie4 = movie2.sort_values(['title_year','imdb_score'], ascending=[False,True]) 
movie4.head()
```
