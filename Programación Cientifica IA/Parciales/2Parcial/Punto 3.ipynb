```
import nltk 
import json
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')

with open('drive/My Drive/7 Semestre/Dataset1Parcial2.json') as file:
  data = json.load(file)
```

```
#Función para generar un n-grama de reviews
def nGrama(n, reviews):
  Ngrama=list()
  for y in reviews:  
      POS = nltk.pos_tag(y, tagset='universal') 
      for i in range(0,len(y)-(n-1)):
        tag = ""
        l = n+i
        for num in range(i, l):
          if num == l-1:
            tag += POS[num][1]
            break
          tag += POS[num][1]+"-"
          
        if(tag not in Ngrama):
          Ngrama.append(tag)

  return Ngrama

nGramUsers = list()

#n-grama de reviews de cada usuario
for pr in data:
  nGramUsers.append(nGrama(2, pr['reviews']))

print(nGramUsers)
```

```
#Eliminar palabras que no sean stop words
from gensim.parsing.preprocessing import remove_stopwords

t1 = dict()
for p in data:
  t1[p['user']] = list()
  for y in p['reviews']:
    rev = remove_stopwords(" ".join(y))
    q = y
    for pal in rev.split():
      if pal in q:
        q.remove(pal)
    t1[p['user']].append(q)
```

```
#Realizar una representación vectorial Tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(nGramUsers, t1).toarray()
features
```
