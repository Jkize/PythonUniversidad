```python
import copy
```

```python
class Estado: 
    #Flecha: None, up, down, left, right
    def __init__(self,row,col):
        self.mat=[]
        for i in range(row):
            self.mat.append(list(range(col)))
        self.obstaculo=0
        self.libre=row*col
        self.row=row
        self.col=col
        self.visitado=0
        
    def addObstaculo(self,x,y):
        self.mat[x][y]="*"
        self.obstaculo+=1
        self.libre-=1
    
    def __str__(self):
        res=""
        for row in self.mat:
            res+="{} \n".format(row)
        return res    
```

```python
def genEstado(x,y,flechaI,mat):
    estados=[]
    flechas=["🡠","🡡","🡣","🡢"]
    flecha=[]
    if(flechaI=="🡢"):
        y+=1
        flechas=["🡢","🡡","🡣"]
    elif(flechaI=="🡠"):
        y-=1
        flechas=["🡠","🡡","🡣"]
    elif(flechaI=="🡡"):
        x-=1
        flechas=["🡢","🡡","🡠"]
    else:
        x+=1
        flechas=["🡢","🡣","🡠"]
    if(x>=0 and x<8 and y>=0 and y<6 and mat[x][y]!="*" and (mat[x][y] not in flechas)):
        for fl in flechas:
            estados.append({"x":x,"y":y,"flecha":fl})
    return estados
```

```python
def DFS(posX,posY,flecha,estado): 
    estado.mat[posX][posY]=flecha
    estado.visitado+=1 
    if(estado.visitado==estado.libre):
        return estado
    estados=genEstado(posX,posY,flecha,estado.mat)    
    for dic in estados: 
        res=DFS(dic["x"],dic["y"],dic["flecha"],copy.deepcopy(estado))
        if(res!=None):
            return res
    return None             
```

```python
inicial=Estado(8,6)
inicial.addObstaculo(0,4)
inicial.addObstaculo(3,2)
inicial.addObstaculo(4,0)
inicial.addObstaculo(4,5)
 
estados=genEstado(0,2,"🡡",inicial.mat)
for i in estados:
    print(i)
res=DFS(0,2,"🡢",inicial)
print(res)

#res=DFS(1,2,"🡡",inicial) 
#print(res)
```
