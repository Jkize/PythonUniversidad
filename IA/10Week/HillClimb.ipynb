```python
import numpy as np
import random
import numpy.random as rn
import copy
#Reina 4x4
globalSize=8
```

```python

class Reina:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return "({}, {})".format(self.x, self.y)
    def __eq__(self,other):
        if(self.x==other.x and self.y==other.y):
            return True
        

class Estado:
    def __init__(self,reinas):
        self.reinas=reinas
    
    def colision(self,r1,r2):
        #Reviso sÃ­ esta en el mismo eje
        if(r1.x==r2.x):
            return True
        if(r1.y==r2.y):
            return True
        valx=abs(r1.x-r2.x)
        valy=abs(r1.y-r2.y)
        #Diagonal
        if(valx==valy):
            return True
        
    def score(self):
        cont=0
        for idx in range(len(self.reinas)-1):
            r1=self.reinas[idx]
            
            for idx2 in range(idx+1,len(self.reinas)):
                r2=self.reinas[idx2]
                if(self.colision(r1,r2)):
                    #print(r1," con ",r2)
                    cont-=1
        return cont
    
    def __str__(self):
        array=np.zeros((globalSize, globalSize)) 
        array = array.astype('int') 
        array=array.astype('str')
        imprimir=""
        for r in self.reinas:
            array[r.x][r.y]="*"
        for t in array: 
            print(t)
        return ""
```

```python

def generateRandomSolution():
    reinas=[]
    for i in range(globalSize):
        reinas.append(Reina(random.randint(0,globalSize-1),random.randint(0,globalSize-1)))
    estado=Estado(reinas)
    diferent(estado)
    return estado
    
def diferent(estado):
    for i in range(globalSize):
        while(True):
            r1=estado.reinas[i]
            bol=False
            for i2 in range(globalSize):
                if(i!=i2):
                    if(r1==estado.reinas[i2]):
                        bol=True
                        break
            if(bol==True):
                mutate(estado,i)
            if(bol==False):
                break
            
        
        

def mutate(estado,idx):
    if(rn.random()>0.5):
        estado.reinas[idx].x=random.randint(0,globalSize-1)
    if(rn.random()>0.5):
        estado.reinas[idx].y=random.randint(0,globalSize-1)
    
def mutateSolution(estado):
    idx=random.randint(0, globalSize-1)
    mutate(estado,idx)
```

```python
best=generateRandomSolution()
bestScore=best.score()
print(best,"Score: ",bestScore,"\n")
iteraciones=0
while True:
    if(iteraciones==1000):
        break
    newSolution=copy.deepcopy(best)
    mutateSolution(newSolution)
    newScore=newSolution.score()
    if(newScore>bestScore):
        bestScore=newScore
        best=newSolution
        print(best,"Score: ",bestScore,"\n")
    iteraciones+=1
```
