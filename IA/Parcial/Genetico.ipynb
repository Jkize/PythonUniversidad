```python
POSICIONES=[[0,0],[0,1],[0,3],[0,5],[1,2],[1,5],
           [2,0],[2,4],[3,1],[3,5],[4,0],[4,3],
           [5,0],[5,2],[5,4],[5,5]]
SUDOKU=[[0,0,3,0,1,0],
       [5,6,0,3,2,0],
       [0,5,4,2,0,3],
       [2,0,6,4,5,0],
       [0,1,2,0,4,5],
       [0,4,0,1,0,0]]   
```

```python
LONGITUD_CROMOSOMA = 16

class Individuo:
    def __init__(self, cromosoma):  
        """
        Parameters:
        cromosoma: List de int de 1s y 0s, cada valor se denomina gen
        """
        self.cromosoma = cromosoma
        self.porcentaje=0 #Para la ruleta
    
    def _array(self):
        for idx,val in enumerate(self.cromosoma):
            SUDOKU[POSICIONES[idx][0]][POSICIONES[idx][1]]=val
        return np.array(SUDOKU)
    
    #Diferentes números
    def _diferente(self,mat):
        _set={0}
        for row in mat:
            for dt in row:
                _set.add(dt)
        if(len(_set)-1==6):
            return True
        else:
            return False
                
    #Función fitness
    def fevaluar(self):
        arr=self._array()
        cont=0 
        f1=0
        f2=2
        for i in range(3):
            c1=0
            c2=3
            for i in range(2):        
                if( not  self._diferente(arr[f1:f2,c1:c2])):
                    cont-=1  
                c1=3
                c2=6
            f1+=2
            f2+=2
        n1=0
        n2=1
        for i in range(6):
            if(not  self._diferente(arr[n1:n2,:])):
                cont-=1
            if(not  self._diferente(arr[:,n1:n2])):
                cont-=1 
            n1+=1
            n2+=1 
        return cont
    
    
    def __gt__(self, otro):
        return self.fevaluar() > otro.fevaluar()
    
    def __eq__(self, otro):
        return self.cromosoma == otro.cromosoma
    
    def __str__(self):
        return (self.cromosoma.__str__() + " puntaje: {}".format(self.fevaluar()))
    
    def __repr__(self):
        return self.__str__()

#i1 = Individuo(cromosoma=[1,2,3,4,5,6, 1, 2,3,4,5,6,1,2,3,4])
#print(i1._array())
```

```python
import random
def mutar(individuo, prob=0.1):
    """
      Recibe un individuo y retorna otro individuo donde
      cada uno los genes de su cromosoma, pudo haberse mutado con probabilidad prob.
      
      Parameters:
      individuo: Individuo
      prob: float, probabilidad de mutación de un gen, valor entre 1 y 6
      
      Returns:
      individuo: Individuo      
    """    
    for index, gen in enumerate(individuo.cromosoma):
        if random.random() < prob:
            ant=individuo.cromosoma[index]
            while True:
                pos=random.randint(1, 6) 
                if(ant!=pos):
                    individuo.cromosoma[index]=pos
                    break
                        
    return individuo

#i1 = Individuo([1,0,1,0,0,0])
#print(mutar(i1))
```

```python
def generar_random_cromosoma(longitud=5):
    """
        Genera un cromosoma de cierta longitud cuyos genes se generan de manera aleatorio.
        
        Parameters:
        longitud: int, longitud del cromosoma (tamano de la lista)
        
        Returns:
        cromosoma: List de int de 1s y 0s
    """
    cromosoma = []
    for i in range(longitud):
        gen = random.randint(1,6)
        cromosoma.append(gen)
    return cromosoma
    
#generar_random_cromosoma()
```

```python
def generar_poblacion_inicial(tamano_poblacion=10, longitud_cromosoma=5):
    """
        Genera una población inicial de individuos donde cada individuo tiene 
        un cromosoma generado de manera aleatoria.
        
        Parameters:
        tamano_poblacion: int, número de individuos a generar
        longitud_cromosoma: int, longitud del cromosoma de cada individuo
        vmin: int, valor mínimo que puede tener el fenotipo de los individuos generados
        vmax: int, valor máximo que puede tener el fenotipo de los individuos generados
        
        Returns:
        poblacion: List[Individuo], lista de individuos
        
    """       
    poblacion = []
    for i in range(tamano_poblacion):
        individuo=Individuo(generar_random_cromosoma(longitud=longitud_cromosoma))
        poblacion.append(individuo) 

    return poblacion

#generar_poblacion_inicial(longitud_cromosoma=LONGITUD_CROMOSOMA)
```

```python
#poblacion = generar_poblacion_inicial(5, LONGITUD_CROMOSOMA)

def seleccionar_padre_por_toreno(poblacion):
    """
        Selecciona un individio de la población
        para ser padre utilizando el método de torneo.
        
        Parameters:
        poblacion: list[Individuo], lista de individuos de 
        donde se va a seleccionar uno como padre
        
        Returns:
        padre: Individuo
    """
    
    padre1 = random.choice(poblacion)
    padre2 = random.choice(poblacion)
    padre = max(padre1, padre2)
    return padre

def seleccionar_padre_por_ruleta(poblacion):
    sumT=sum(list(map(lambda dt:dt.fevaluar(),poblacion)))
    for i in range(len(poblacion)):
        poblacion[i].porcentaje=1-((poblacion[i].fevaluar())/sumT)
    poblacion.sort(key=lambda x: x.porcentaje, reverse=True)
    ra=random.random()
    raT=0
    for dt in poblacion:
        raT+=dt.porcentaje
        if(ra<=raT):
            return dt
    

#print(poblacion)
#print("\n")
#p = seleccionar_padre_por_toreno(poblacion)
#print(p)
```

```python
import copy
poblacion = generar_poblacion_inicial(20, LONGITUD_CROMOSOMA)
padres = []
def generar_padres(poblacion, num_padres=6,isTorneo=True):
    """
        Genera una lista de padres a partir de una población(lista de individuos).
        Utiliza el método de selección por torneo.
        
        Parameters:
        poblacion: List[individuo]
        num_padres: int, número de padres a seleccionar
        
        Returns:
        padres: list[Individuo]
    """
    
    posibles_padres = copy.deepcopy(poblacion)
    padres_seleccionados = []
    posibles_padres.sort(reverse=True)
    for i in range(num_padres):
        padre=None
        if(isTorneo):
            padre = seleccionar_padre_por_toreno(posibles_padres)
        else:
            padre = seleccionar_padre_por_ruleta(posibles_padres)        
        padres_seleccionados.append(padre)
        posibles_padres.remove(padre)
        
    return padres_seleccionados
#padres = generar_padres(poblacion)
#print(poblacion,"\n")
#print(padres)
```

```python
def runGenetico(initPoblacion,prob_mutacion=0.1,isTorneo=True):
    poblacion=copy.deepcopy(initPoblacion)
    listScore=[]
    anterior_mejor = None
    G=0
    for i in range(150):
        G=i
        poblacion.sort(reverse=True)
        anterior_mejor = poblacion[0]
        nueva_poblacion = []
        scoreAnteriorMejor=anterior_mejor.fevaluar()
        listScore.append(scoreAnteriorMejor)
        if(scoreAnteriorMejor==0):
            break
        
        for i in range(int(len(poblacion)/2)):
            padres =generar_padres(poblacion,2,isTorneo)
            hijo1 = cruzar(padres[0],padres[1])
            hijo2 = cruzar(padres[1],padres[0])
            ## mutar hijos
            hijo1 = mutar(hijo1,prob_mutacion)
            hijo2 = mutar(hijo2,prob_mutacion)
            ## agrega hijos a la nueva población
            nueva_poblacion.append(hijo1)
            nueva_poblacion.append(hijo2)
            ## ...
        
        poblacion = nueva_poblacion
        
        # Eltismo: Incluye al mejor de la generación anterior en la nueva generación
        poblacion.sort(reverse=True)
        poblacion[-1] = anterior_mejor
        
    print("G{}".format(G),anterior_mejor)
    return anterior_mejor,np.array(listScore)  
```

```python
import matplotlib.pyplot as plt
## ALGORITMO GENÉTICO 
#Coger la misma población para las diferentes variaciones
poblacion = generar_poblacion_inicial(10, LONGITUD_CROMOSOMA)
#Con torneo
solT,gT=runGenetico(poblacion,prob_mutacion=0.1,isTorneo=True)
#Con ruleta
solR,gR=runGenetico(poblacion,prob_mutacion=0.1,isTorneo=False)
#Con Torneo 0.2 Prob
solT_2,gT_2=runGenetico(poblacion,prob_mutacion=0.2,isTorneo=True)
#Con ruleta 0.2 proba
solR_2,gR_2=runGenetico(poblacion,prob_mutacion=0.2,isTorneo=False)
#Con Torneo 0.5 Prob
solT_5,gT_5=runGenetico(poblacion,prob_mutacion=0.5,isTorneo=True)
#Con ruleta 0.5 proba
solR_5,gR_5=runGenetico(poblacion,prob_mutacion=0.5,isTorneo=False)
```
